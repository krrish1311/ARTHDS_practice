# Image

```python
img=np.zeros((480,640,3),dtype=np.uint8)
```

```python
img1=img.copy()
```

```python
plt.imshow(img1)
```

```python
img1=cv2.rectangle(img1,(0,200),(400,300),(255,255,255),-1)
```

```python
plt.imshow(img1)
```

```python
img2=img.copy()
img2=cv2.rectangle(img2,(200,200),(600,300),(255,255,255),-1)
```

```python
plt.imshow(img2)
```

# AND

```python
img_and=cv2.bitwise_and(src1=img1,src2=img2)
```

```python
plt.imshow(img_and)
```

# OR

```python
img_or=cv2.bitwise_or(src1=img1,src2=img2)
```

```python
plt.imshow(img_or)
```

# XOR

```python
img_xor=cv2.bitwise_xor(src1=img1,src2=img2)
```

```python
plt.imshow(img_xor)
```
