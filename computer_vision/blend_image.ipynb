# in blending wE ADD THE PHOTO COLORS IN SOME PROPORTIONS WHERE  we first take the alpha% of first image color 
# and then add it with the beta% of second pic and then the added colors of each pixels creates the blanded image 

# we have R,G,B colors for each 3d image .in blending it first take R of 1st pixel of pic1 multiply it with alpha R*alpha and then
# it takes the R of 1st pixel of 2nd pic and add it by R*beta so R*alpha + R*beta .similarly for green and blue 
# G*alpha + G*beta ,B*alpha + B*beta . here we get first pixel [R*alpha + R*beta,G*alpha + G*beta ,B*alpha + B*beta]

```python
pic2=cv2.resize(pic2,(1920,1080))
```

```python
#in blending wE ADD THE PHOTO COLORS IN SOME PROPORTIONS WHERE  we first take the alpha% of first image color 
#and then add it with the beta% of second pic and then the added colors of each pixels creates the blanded image 

#we have R,G,B colors for each 3d image .in blending it first take R of 1st pixel of pic1 multiply it with alpha R*alpha and then
#it takes the R of 1st pixel of 2nd pic and add it by R*beta so R*alpha + R*beta .similarly for green and blue 
#G*alpha + G*beta ,B*alpha + B*beta . here we get first pixel[R*alpha + R*beta,G*alpha + G*beta ,B*alpha + B*beta]
```

```python
b1,r1,g1=cv2.split(pic1)
```

```python
b2,r2,g2=cv2.split(pic2)
```

```python
row,column=b1.shape
```

```python
pic1.shape
```

```python
for i in range(row):
    for j in range(column):
        b3[i,j]=b1[i,j]*0.8 + b2[i,j]*0.2
```

```python
import numpy as np
```

```python
b3=np.zeros((1080,1920),dtype=np.uint8)
```

```python
r3=np.zeros((1080,1920),dtype=np.uint8)
```

```python
for i in range(row):
    for j in range(column):
        r3[i,j]=g1[i,j]*0.8 + g2[i,j]*0.2
```

```python
blend=np.zeros((1080,1920,3),dtype=np.uint8)
```

```python
blend[:,:,0]=b3
blend[:,:,1]=g3
blend[:,:,2]=r3
```

```python
blend=cv2.cvtColor(blend,cv2.COLOR_BGR2RGB)
```
